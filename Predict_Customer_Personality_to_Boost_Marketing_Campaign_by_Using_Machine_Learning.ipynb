Feature Engineering : Conversion Rate

```python
df['total_transaction'] = df['NumDealsPurchases'] + df['NumWebPurchases'] + df['NumCatalogPurchases'] + df['NumStorePurchases']

def div(x,y):
    if y == 0:
        return 0
    return x / y

df['conversion_rate'] = df.apply(lambda x : div(x['total_transaction'], x['NumWebVisitsMonth']), axis=1)
```

Age Group

```python
df['age'] = 2022 - df['Year_Birth']

df['age_group'] = np.where((df['age'] >= 0) & (df['age'] <=40), 'young adult',
                             (np.where((df['age'] >= 41) & (df['age'] <= 55), 'adult',
                             (np.where((df['age'] >= 56) & (df['age'] <= 65), 'middle aged', 'senior')))))
df
```

```python
df['total_response'] = df['AcceptedCmp3'] + df['AcceptedCmp5'] + df['AcceptedCmp4'] + df['AcceptedCmp2'] + df['AcceptedCmp1'] + df['Response']
df['total_spent'] = df['MntCoke'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']
df["children"] = df["Kidhome"] +  df["Teenhome"]
df['has_partner'] = df['Marital_Status'].replace({'Lajang' : 'alone', 'Menikah' : 'partner', 'Janda' : 'alone', 'Duda' : 'alone', 'Cerai' : 'alone', 'Bertunangan' : 'partner'})
df["is_parent"] = np.where(df.children > 0, 1, 0)
df['Education'] = df['Education'].replace({'S1' : 'sarjana', 'S2' : 'sarjana', 'S3' : 'sarjana', 'D3' : 'diploma', 'SMA' : 'undergraduate'})
```

```python
## drop data null
df.dropna(inplace = True)
df.isna().sum()
```

```python
# Check data duplicated
df.duplicated().sum()
```

```python
# Handling Outliers
nums = ['Income','age','total_transaction']
features = nums
plt.figure(figsize=(10, 8))
for i in range(0, len(features)):
    plt.subplot(1, len(features), i+1)
    sns.boxplot(y=df[features[i]], color='darkcyan', orient='v')
    plt.tight_layout()
```

```python
num_col = ['Income','age','total_transaction']

for col in num_col:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3-q1
    ll = q1-(1.5*iqr)
    ul = q3+(1.5*iqr)
    for ind in df[col].index:
        if df.loc[ind,col]>ul:
            df.loc[ind,col]=ul
        elif df.loc[ind,col]<ll:
            df.loc[ind,col]=ll
        else:
            pass
print('Handling outliers finished')
```

DATA MODELLING

```python
# selecting 3 components 
from sklearn.decomposition import PCA

pca=PCA(n_components=3)
pca.fit(df_std)
pca_df = pd.DataFrame(pca.transform(df_std), columns=(['PC1','PC2','PC3']))
```

```python
# Elbow Method
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer

visualizer = KElbowVisualizer(KMeans(), k=(10))
visualizer.fit(pca_df)    
visualizer.show();    
```

```python
from sklearn.metrics import silhouette_score

# kmeans = KMeans(4)
# km = kmeans.fit(pca_df)  
# KM_Labels = km.labels_
# pca_df['cluster'] = kmeans.fit_predict(pca_df)
# pca_df 

km = KMeans(n_clusters=4, init='k-means++', n_init=10, max_iter=100, random_state=42)
km.fit_predict(pca_df)
# Add labels to df 
pca_df['Cluster_Id'] = km.labels_
df['Cluster_Id'] = km.labels_
df.info()
```

```python
from yellowbrick.cluster import SilhouetteVisualizer
from sklearn.metrics import silhouette_samples, silhouette_score

fig, ax = plt.subplots(2, 2, figsize=(15,8))
for i in [2, 3, 4, 5]:
    '''
    Create KMeans instance for different number of clusters
    '''
    km = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=100, random_state=42)
    q, mod = divmod(i, 2)
    '''
    Create SilhouetteVisualizer instance with KMeans instance
    Fit the visualizer
    '''

    visualizer = SilhouetteVisualizer(km, colors='yellowbrick', ax=ax[q-1][mod])
    visualizer.fit(pca_df)
    
    km.fit_predict(pca_df)
    #
    # Calculate Silhoutte Score
    #
    score = silhouette_score(pca_df, km.labels_, metric='euclidean')
    #
    # Print the score
    #
    print(f'Silhouette Score: for {i} clusters is {score}')
```
